```python
import openpyxl

# Load the Excel file using openpyxl
workbook = openpyxl.load_workbook('Match_List_WBBL.xlsx')

# Assuming 'Sheet1' is the name of the sheet containing the hyperlinks
sheet = workbook['Sheet1']

# Specify the column and rows where hyperlinks are located
column_letter = 'G'  # Replace with the actual column letter
start_row = 2  # Replace with the actual starting row

# Extract hyperlinks from the specified column
hyperlinks = [cell.hyperlink.target for cell in sheet[column_letter] if cell.hyperlink]

# Display the extracted hyperlinks
print(hyperlinks)
```

```python
df['hyperlinks'] = hyperlinks
```

```python
# Filtering out records of matches where result was not determined
df_subset = df[df['Winner']!='no result']
```

```python
df_subset.info()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


# Function to scrape data from a given URL
def scrape_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extracting data based on provided classes
    Season_Details =soup.find_all('span', class_='ds-text-tight-s ds-font-medium ds-text-typo ds-underline ds-decoration-ui-stroke hover:ds-text-typo-primary hover:ds-decoration-ui-stroke-primary ds-block')[2]
    Team_Batting_First = soup.find('span', class_='ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate')
    Team_Batting_Second = soup.find_all('span', class_='ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate')[1]
    Team1_Score = soup.find('div', class_='ds-text-compact-m ds-text-typo ds-text-right ds-whitespace-nowrap').strong
    Team2_Score = soup.find_all('div', class_='ds-text-compact-m ds-text-typo ds-text-right ds-whitespace-nowrap')[1].strong
    Full_Result_Detail= soup.find('p',class_='ds-text-tight-m ds-font-regular ds-truncate ds-text-typo')
    #print(Full_Result_Detail)
    # Extracting text content from BeautifulSoup objects
    
    Season_Details_text = Season_Details.text if Season_Details else None
    Team_Batting_First_text = Team_Batting_First.text if Team_Batting_First else None
    Team_Batting_Second_text = Team_Batting_Second.text if Team_Batting_Second else None
    Team1_Score_text = Team1_Score.text if Team1_Score else None
    Team2_Score_text = Team2_Score.text if Team2_Score else None
    Full_Result_Detail_text= Full_Result_Detail.get_text(strip=True) if Full_Result_Detail else None
          
    return Season_Details_text, Team_Batting_First_text, Team_Batting_Second_text, Team1_Score_text, Team2_Score_text, Full_Result_Detail_text

# Scraping data for each hyperlink
data = []
for link in df_subset['hyperlinks']:
    result = scrape_data(link)
    data.append(result)
    print(len(data))

```

```python
# Creating a DataFrame
columns = ['Match_Details', 'Team_Batting_First', 'Team_Batting_Second', 'Team1_Score','Team2_Score', 'Result_Details']
df1 = pd.DataFrame(data, columns=columns)

# Displaying the DataFrame
print(df1)
```

```python
## Combining the two dataframes for further analysis
result_df = pd.concat([df_subset.reset_index(drop=True), df1.reset_index(drop=True)], axis=1)
## Important to reset the index as the subsets usually inherit the original index.
```
